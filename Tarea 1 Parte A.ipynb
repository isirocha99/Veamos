## Pregunta 1

````python
# Escriba su código aquí
#metros plaza: df_datos_comunas población: df_población
df1 = df_datos_comunas.merge(df_poblacion, left_on= 'CODIGO', right_on= 'CODIGO')
df1= df_codigos.merge(df1, left_on='CODIGO', right_on='CODIGO')
df1.fillna(0)
df1['metrosplaza / poblacion'] = df1['metros_plaza'] / df1['POBLACIÓN']
df1 = df1[['Nombre Comuna', 'metrosplaza / poblacion']]
df1.sort_values(by='metrosplaza / poblacion', ascending=False)
````

## Pregunta 2

````python
# Escriba su código aquí
df2 = df_codigos.merge(df_presupuesto_2019, left_on= 'CODIGO', right_on= 'CODIGO')
df2 = df2[['Nombre Región', 'PRESUPUESTO']]
df2 = df2['PRESUPUESTO'].groupby([df2['Nombre Región']]).mean()
print('La región de mayor presupuesto es la Región Metropolitana de Santiago')
df2.sort_values(ascending=False)
````
